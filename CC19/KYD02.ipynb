```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

pokemonsDf = pd.read_csv('Pokemon.csv')

# ((pokemonsDf['Type 1'] == "Normal") & (pokemonsDf['Type 2'] == "Fighting")) | ((pokemonsDf['Type 1'] == "Normal") & (pokemonsDf['Type 2'] == "Fighting"))  
xd = pokemonsDf[((pokemonsDf['Type 1'] == "Normal") | (pokemonsDf['Type 2'] == "Normal")) | ((pokemonsDf['Type 1'] == "Fighting") | (pokemonsDf['Type 2'] == "Fighting"))]
xd
```
